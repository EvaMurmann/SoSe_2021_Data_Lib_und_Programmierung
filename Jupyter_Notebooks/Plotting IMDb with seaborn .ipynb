```python
# Ein zufälliges Subset aus der Gesamtmenge extrahieren
# random_state ist eine beliebiger Startwert
imdb_subsample = imdb.sample(10000, random_state=42)
```

```python
imdb.shape
```

```python
sns.stripplot(x=imdb_subsample.averageRating)
```

```python
sns.boxenplot(x=imdb_subsample.averageRating)
sns.stripplot(x=imdb_subsample.averageRating, alpha=0.1, color=("orange"))
```

```python
sns.swarmplot(y=imdb_subsample.sample(500, random_state=1) .averageRating)
```

```python
sns.scatterplot(x=imdb.averageRating, y=imdb.numVotes, color=("green"))
```

```python
sns.scatterplot(data=imdb, x="averageRating", y="numVotes")
```

```python
# mit der Methoder apply, kann eine Funktion auf alle Werte einer Serie angewandt werden;
# es wird eine neue Serie zurückgegeben, die die Rückgabewerte der Funktion enthält 
def multiply_by_1000(number):
    return "Blub-" + str (number * 1000)
imdb.averageRating.apply(multiply_by_1000)
```

```python
# Lambda Funktionen (Unbenannte Funktionen )
imdb.averageRating.apply(lambda number:"Blub-" + str (number * 1000))
```

```python
# Wir generieren eine neue Spalte, in der wir unsere Filme in zwei KLassen einteilen. 
# Solche die 20 (der Median) oder mehr Stimmen bekommen haben und solche 
# die weniger Stimmen bekommen haben
imdb_subsample["voting_class"] = imdb_subsample.numVotes.apply(
    lambda number: "more_or_equal_20" if number >= 20 else "less_than_20"
)
```

```python
imdb_subsample
```

```python
sns.stripplot(data=imdb_subsample, y="averageRating", x="voting_class", alpha=0.1)
```

```python
sns.boxplot(data=imdb_subsample, y="averageRating", x="voting_class")
```
